# Data pre-processing

### Import pandas & Load csv

```python
import pandas as pd
df = pd.read_csv('player_info_ArsenalFC.csv')
df.head()
```

### Information of Data

```python
df.shape # (rows, columns)
```

```python
df.info
```

### Data pre-processing

```python
df['value'] = df['value'].str.replace('€','')
df['value'] = df['value'].str.replace('m','')
df['value'] = df['value'].str.replace('300Th.','0.3')
df['value'] = df['value'].astype('float')
df.head()
```

```python
df['시장가치(억)'] = df['value']*13.42  # Change the currency : € => \
df.head()
```

```python
df.drop(columns = ['value'],inplace=True)
df.head()
```

```python
df.to_csv('1_player_info_ArsenalFC.csv')
```
