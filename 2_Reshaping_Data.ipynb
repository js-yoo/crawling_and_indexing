# Reshaping Data

### Import pandas & Load csv

```python
import pandas as pd
df1 = pd.read_csv('1_player_info_ArsenalFC.csv')
df1.drop(columns = ['Unnamed: 0'], inplace=True)
df1.head()
```

## Reshaping Data

### Order rows by values of a column (low to high)

```python
df1.sort_values('시장가치(억)').head()
```

### Order rows by values of a column(high to low)

```python
df1.sort_values('시장가치(억)',ascending=False).head()
```

### Rename the columns of a DataFrame

```python
df1.rename(columns = {'age':'birth (age)'},inplace = True)
df1.to_csv('2_player_info_ArsenalFC.csv')
df2 = pd.read_csv('2_player_info_ArsenalFC.csv')
df2.drop(columns = ['Unnamed: 0'], inplace=True)
df2.head()
```

### Sort the index of a DataFrame

```python
df2.sort_index().head()
```

```python
df2.sort_index(ascending = False).head()
```

### Reset index of DataFrame to row numbers, moving index to columns

```python
df2.reset_index().head()
```

### Drop columns from DataFrame

```python
df2.drop(columns = ['birth (age)', 'nation']).head()
```
