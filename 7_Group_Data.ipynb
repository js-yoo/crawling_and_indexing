# Group Data

### Import pandas & Load csv

```python
import pandas as pd
df2 = pd.read_csv('2_player_info_ArsenalFC.csv')
df2.drop(columns = ['Unnamed: 0'], inplace=True)
```

### Group and Analyze data

```python
df2.groupby('position')
```

```python
g = df2.groupby('position')
g.size()
```

```python
g.count()
```

```python
g['시장가치(억)'].sum()
```

```python
c = df2.groupby('nation')
c.size()
```

```python
c['시장가치(억)'].sum().sort_values(ascending = False)
```

### Ranks. Toes get min rank

```python
df2.rank(method='min')
```
