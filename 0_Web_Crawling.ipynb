# Bring Arsenal Player info.

## 1. Using Web Crawling(BeautifulSoup & requests)

### Import required Library

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

### Create empty List for plyer's info.

```python
number = []   # this means Back # or Uniform #
name = []
position = []
age = []      # Date of birth/Age ex. Aug 6, 1997(25)
nation = []   # Dual citizenship isn'y considered
value = []
```

### Bring player info. using 'requesets'

```python
# headers = {'User-Agent' : 'User info'} To inform that you are not a robot
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.104 Safari/537.36'}
# url = 'Address of the website'
# Before use, 'url/rovots.txt' [allow] must be checked
url="https://www.transfermarkt.com/arsenal-fc/startseite/verein/11"
# Make a request with request.get()
r=requests.get(url, headers=headers)
```

### Check the response (200 = OK, 404 = Not Found)

```python
r.status_code
```

## 2. Saving as .csv file

### Preparing to analyze imported data

```python
soup = BeautifulSoup(r.content, 'html.parser') # either r.content or r.text
# Find a tag and class with player info. ,
# tag : 'tr' , class : 'odd', 'even'
player_info = soup.find_all('tr',class_=['odd','even'])
```

### Append info. to each List

```python
for info in player_info:
    player = info.find_all('td') # 'td' tag has each player's info
    number.append(player[0].text)
    name.append(player[2].img['alt'])
    position.append(player[4].text)
    age.append(player[6].text)
    nation.append(player[7].img['alt'])
    value.append(player[8].text)
```

### Save as DataFrame

```python
df=pd.DataFrame(
        {"number":number,
         "name":name,
         "position":position,
         "age":age,
         "nation":nation,
         "value":value   
        })
```

### Check the result

```python
df
```

### Save as .csv

```python
df.to_csv('player_info_ArsenalFC.csv', index=False)
```
